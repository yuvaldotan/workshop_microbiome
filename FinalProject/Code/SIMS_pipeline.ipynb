```python
baboon_pairs = [np.random.choice(list(all_model.baboons.values()), 2,replace=False) for i in range(10000)]
bray_curtis = []
for i in range(10000):
    baboon1, baboon2 = baboon_pairs[i]
    baboon1_samples = baboon1.data_I.index
    baboon2_samples = baboon2.data_I.index
    baboon1_data = baboon1.data_I.loc[np.random.choice(baboon1_samples, 1)[0]].values
    baboon2_data = baboon2.data_I.loc[np.random.choice(baboon2_samples, 1)[0]].values
    bc = braycurtis(baboon1_data, baboon2_data)
    bray_curtis.append(bc)
```

```python
mean_bc = np.mean(bray_curtis)
sns.kdeplot(bray_curtis)
plt.vlines(mean_bc, 0, 2.5, color='red', linestyle='--')
plt.title(f"Bray-Curtis Distribution:\nrandom pairs of baboons\nmean={mean_bc:.3f}")
plt.show()
```
